```python
import os
from dotenv import load_dotenv #use to load environment variables from a .env file
from langchain_community.document_loaders import PyPDFLoader #use to load pdf files
from langchain_text_splitters import RecursiveCharacterTextSplitter #use to split text into smaller chunks
from langchain_huggingface import HuggingFaceEmbeddings #use to generate embeddings for text using HuggingFace models
from langchain_community.vectorstores import FAISS
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_core.messages import HumanMessage
from langchain_core.messages import SystemMessage
from langchain_core.messages import AIMessage
```

```python
from langchain_huggingface import HuggingFaceEndpoint
from langchain_huggingface import ChatHuggingFace
#HuggingFaceEndpoint is used to connect to a HuggingFace model hosted on the HuggingFace Hub. It allows you to send requests to the model and receive responses.
llm = HuggingFaceEndpoint(repo_id="meta-llama/Llama-3.1-8B-Instruct",
    max_new_tokens=512,
    temperature=0.5,
    huggingfacehub_api_token=os.getenv("HF_TOKEN"))
chat_model = ChatHuggingFace(llm=llm)



# ChatHuggingFace (Chat Model)
# - Wraps an LLM to provide a chat interface.
# - Works with chat messages (SystemMessage, HumanMessage, AIMessage).
# - Returns an AIMessage object instead of a plain string.
# - Designed for chatbots, conversational AI, memory, and multi-turn conversations.
# - Does NOT automatically store chat history; history must be managed separately (e.g., Memory or RunnableWithMessageHistory).
```

### **Prompt template**

```python
question = "What is Machine Learning?"

prompt = f"""
You are an AI teacher.

Answer the following question clearly.

Question:
{question}

Answer:
"""

response = chat_model.invoke(prompt)
print(response.content)


# Why do we need PromptTemplate?
#
# Without PromptTemplate:
# - We have to write the complete prompt every time.
# - The same prompt structure gets repeated in multiple places (code duplication).
# - If the prompt needs to be updated, we must edit it everywhere.
# - Dynamic values (e.g., question, context, language) have to be inserted manually using f-strings or string concatenation.
# - Prompts become difficult to maintain as the project grows.
#
# PromptTemplate solves this by:
# - Defining the prompt structure once.
# - Replacing only the variables (placeholders) at runtime.
# - Making prompts reusable, cleaner, and easier to maintain.
```

#### **Using Prompt Template**

```python
# Step 1: Create the Prompt Template
# Define a reusable prompt structure with placeholders (variables).
# At this stage, no actual values are provided.


from langchain_core.prompts import PromptTemplate
prompt = PromptTemplate.from_template("""
You are a {profession} expert.

Explain the following topic.

Topic:
{question}

Language:
{language}
""")
```

```python
# Step 2: Fill the Prompt Variables
# Replace the placeholders with actual values to generate the final prompt.
formatted_prompt = prompt.invoke(
{
    "profession":"Data Science",

    "question":"Machine Learning",

    "language":"Hindi"
}
)
```

```python
# Step 3: Send the Final Prompt to the LLM
# Pass the formatted prompt to the model and generate a response.
chat_response = chat_model.invoke(formatted_prompt)
print(chat_response.content)
```

### **ChatPrompt Template With Message History**

```python
from langchain_core.prompts import (ChatPromptTemplate,MessagesPlaceholder)
prompt = ChatPromptTemplate.from_messages([("system", "You are an AI Teacher."),
                                           MessagesPlaceholder("history"),("human", "{question}")])
```

```python
from langchain_core.messages import (HumanMessage,AIMessage)

history = [
    HumanMessage(content="What is Machine Learning?"),
    AIMessage(content="Machine Learning is a subset of AI."),
    HumanMessage(content="What is Deep Learning?"),
    AIMessage(content="Deep Learning is a subset of Machine Learning.")
]
```

```python
formatted_prompt = prompt.invoke({"history": history,"question": "Difference between ML and DL?"})
print(formatted_prompt)
```

```python
response= chat_model.invoke(formatted_prompt.messages)
print(response.content)
```

### **Connect to Chain**

```python
chain = prompt | chat_model

response = chain.invoke({
    "history": history,
    "question": "Difference between ML and DL?"
})
```

```python
print(response.content)
```
